```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
import pandas as pd
import pickle 
```

```python
data = pd.read_csv("Used_Bikes.csv")
```

```python
data.dropna(inplace=True)
data.drop_duplicates(inplace=True)
```

```python
data.drop(["bike_name","city"], axis=1, inplace=True)
```

```python
x = data.drop("price", axis=1)
y = data["price"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
ohe = OneHotEncoder()
ohe.fit(data[["owner","brand"]])
```

```python
transposed_columns = make_column_transformer(
    (OneHotEncoder(categories=ohe.categories_), ["owner","brand"]),
    remainder="passthrough",
)
```

```python
model = LinearRegression()
```

```python
pipe = make_pipeline(transposed_columns, model)
```

```python
pipe.fit(x_train, y_train)
```

```python
y_pred = pipe.predict(x_test)
```

```python
pipe.score(x_test,y_test)
```

```python
pipe.score(x_train,y_train)
```

```python
pickle.dump(pipe, open("model.pkl", "wb"))
```
