```python
# logistic regression.py
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
data = pd.read_csv("cust_satisfaction.csv")
data.head()
```

```python
print(data.info())
print(data.isna().sum())
data.dropna(inplace=True)
```

```python

data["Customer Type"].value_counts()
data["Arrival Delay in Minutes"].fillna(0, inplace=True)
data = data.drop_duplicates()
```

```python
loyal_customers = data[data["Customer Type"] == "Loyal Customer"]
disloyal_customers =data[data["Customer Type"] == "disloyal Customer"]
```

```python
loyal_customers = loyal_customers.sample(20000,random_state=42)
balance_pf=pd.concat([loyal_customers,disloyal_customers], axis=0)
balance_pf.reset_index(drop=True,inplace=True)
```

```python
categorical_features = balance_pf.select_dtypes(include='object')
numerical_features = balance_pf.select_dtypes(exclude='object')
```

```python
categorical_cols = categorical_features.columns.tolist()
print(categorical_cols)
categorical_dict = {}
for col in categorical_cols:
    categorical_dict[col] = {val: i for i, val in enumerate(data[col].unique())}
    print(categorical_dict[col])
    categorical_features[col] = data[col].map(categorical_dict[col])
```

```python
final_data = pd.concat([numerical_features, categorical_features], axis=1)
final_data.dropna(inplace=True)
final_data.to_csv("encoded_data2.csv", index=False)
```

```python
# ## Splitting the data
x = final_data.drop("Customer Type",axis=1)
y = final_data["Customer Type"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=25)
# ## Logistic Regression
model= LogisticRegression(max_iter=10000)
model.fit(x_train, y_train)
```

```python
model.score(x_train, y_train)
# model.score(x_test, y_test)
```
