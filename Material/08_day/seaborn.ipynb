```python
sns.boxplot(df["total_bill"])
```

```python
df = pd.read_csv("tips.csv")
q1=df["total_bill"].quantile(0.25)
q3=df["total_bill"].quantile(0.75)
iqr = q3-q1
print(f"q1 = {q1},q3={q3},iqr={iqr}")
lower_bound = q1-(1.5*iqr)
upper_bound = q3-(1.5*iqr)
df["total_bill"] = df["total_bill"]>upper_bound
df["total_bill"] = df["total_bill"]<lower_bound
sns.boxplot(df["total_bill"])
```

```python
sns.displot(df['total_bill'],kde=True)
```

```python
sns.jointplot(x=df["total_bill"],y=df["tip"],kind="hex")
```

```python
sns.pairplot(df)
```

```python
sns.relplot(x=df["total_bill"],y=df["tip"],hue=df["time"],style=df["size"],col=df["day"])
```
