```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("laptopPrice.csv")
data.dropna(inplace=True)
data.duplicated().sum()
data.drop_duplicates(inplace=True)
data.reset_index(drop=True, inplace=True)
data.duplicated().sum()
data.isna().sum()
```

```python
# making plots with pandas 
data['brand'].value_counts().plot(kind='bar', figsize=(15, 5), color='orange')
plt.title('Brand Count')
plt.xlabel('Brand')
plt.ylabel('Count')
plt.show()
```

```python
## making plots with matplotlib
plt.scatter(data['ram_gb'], data['Price'])
plt.title("Price vs RAM")
plt.xlabel("RAM in GB")
plt.ylabel("Price in INR")
plt.show()
```

```python
## customizing the plot with seaborn
plt.figure(figsize=(15, 5))
sns.barplot(x=data['brand'], y=data['Price'], palette='magma')
plt.title("Brand vs Price")
plt.xlabel("Brand")
plt.ylabel("Price in INR")
plt.xticks(rotation=30)
plt.show()
```

```python
## weight vs brand and price in single plot
plt.figure(figsize=(15, 8))
sns.barplot(x=data["Price"], y=data['weight'], hue=data['brand'])
plt.title("Brand vs Weight and Price")
plt.title("Brand vs Weight")
plt.xlabel("Brand")
plt.ylabel("Weight")
plt.xticks(rotation=45)
plt.show()
```

```python
## processor_brand vs price
plt.figure(figsize=(15, 5))
plt.cycler(color=['blue', 'orange', 'green', 'red'])
sns.barplot(x=data['processor_brand'], y=data['Price'],hue=data['brand'])
colors = plt.rcParams['axes.prop_cycle'].by_key()['color']
# Customizing the default color cycle
plt.show()
```

```python
## histogram of price
plt.figure(figsize=(15, 5))
plt.hist(data['Price'], bins=30, color='purple', edgecolor='black')
plt.title("Price Distribution")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()
```

```python
## price vs rating
plt.figure(figsize=(15, 5))
plt.scatter(data['rating'], data['Price'], color='green', alpha=0.5)
plt.title("Price vs Rating")
plt.xlabel("Price")
plt.ylabel("Rating")
plt.show()
```

```python
## price vs rating with regression line
## ratings = 1 star , 2 stars , 3 stars 
# extracting numeric value from it
data['rating'] = data['rating'].str.extract('(\d+)').astype(float)
plt.figure(figsize=(15, 5))
sns.regplot(x=data['rating'], y=data['Price'], color='purple')
plt.title("Price vs Rating with Regression Line")
plt.xlabel("Price")
plt.ylabel("Rating")
plt.show()
```

```python
## processor genth vs price line plot
plt.figure(figsize=(15, 5))
sns.lineplot(x=data['processor_gnrtn'], y=data['Price'], color='purple')
plt.title("Processor Generation vs Price")
plt.xlabel("Processor Generation")
plt.ylabel("Price")
plt.show()
```

```python
plt.figure(figsize=(15, 5))
sns.lineplot(x=data['processor_name'], y=data['Price'], color='purple')
plt.title("Processor Generation vs Price")
plt.xlabel("Processor Generation")
plt.ylabel("Price")
plt.show()
```

```python
### brand and processor name vs price
plt.figure(figsize=(15, 5))
sns.barplot(x=data['processor_name'], y=data['Price'], hue=data['brand'])
plt.title("Brand and Processor Name vs Price")
plt.xlabel("Brand")
plt.ylabel("Price")
plt.xticks(rotation=45)
plt.show()
```

```python
## pie chart of brand
explode_variable = data['brand'].value_counts().index
plt.pie(data['brand'].value_counts(), labels=data['brand'].value_counts().index, autopct='%1.1f%%', startangle=140,explode=(0.03,0,0,0,0,0,0,0))
plt.title("Brand Distribution")
plt.axis('equal') 
plt.show()
```

```python
# subplot of brand and processor name vs price
plt.figure(figsize=(15, 5))
plt.subplot(1, 2, 1)
plt.bar(data['brand'], data['Price'], color='purple')
plt.title("Brand vs Price")
plt.xlabel("Brand")
plt.ylabel("Price")
plt.xticks(rotation=45)
plt.subplot(1, 2, 2)
plt.bar(data['processor_name'], data['Price'], color='orange')
plt.title("Processor Name vs Price")
plt.xlabel("Processor Name")
plt.ylabel("Price")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
